# FPKM-based Gene Expression Using Cufflinks Package

To perform FPKM based gene expression, we use Cufflinks with the
following command:

```python
!cufflinks -p 2 -o brain1 --library-type fr-firststrand --GTF gtf/chr22.gtf --frag-lenmean \
262 --frag-len-std-dev 80 --no-update-check star_SAM/brain1.bam

!cufflinks -p 2 -o brain2 --library-type fr-firststrand --GTF gtf/chr22.gtf --frag-lenmean \
262 --frag-len-std-dev 80 --no-update-check star_SAM/brain2.bam

!cufflinks -p 2 -o uhr1 --library-type fr-firststrand --GTF gtf/chr22.gtf --frag-len-mean \
262 --frag-len-std-dev 80 --no-update-check star_SAM/uhr1.bam

!cufflinks -p 2 -o uhr2 --library-type fr-firststrand --GTF gtf/chr22.gtf --frag-len-mean \
262 --frag-len-std-dev 80 --no-update-check star_SAM/uhr2.bam
```

#Merge Cufflinks Assembly Using Cuffmerge
We will need to join the transcript assembly feature files generated
by Cufflinks to create a single assembly feature file. To do this, use
the following commands:

```python
!ls */transcripts.gtf > assembly_GTF_list.txt
!cuffmerge -p 2 -o merged -g gtf/chr22.gtf -s genome/chr22.fa assembly_GTF_list.txt
```

#Quantify Expression Using Cuffquant
Expressions are then quantified using Cuffquant with the following
commands:

```python
!cuffquant -p 2 --library-type fr-firststrand --frag-len-mean 262 --frag-len-std-dev \
80 --no-update-check -o brain1 merged/merged.gtf star_SAM/brain1.bam

!cuffquant -p 2 --library-type fr-firststrand --frag-len-mean 262 --frag-len-std-dev \
80 --no-update-check -o brain2 merged/merged.gtf star_SAM/brain2.bam

!cuffquant -p 2 --library-type fr-firststrand --frag-len-mean 262 --frag-len-std-dev \
80 --no-update-check -o uhr1 merged/merged.gtf star_SAM/uhr1.bam

!cuffquant -p 2 --library-type fr-firststrand --frag-len-mean 262 --frag-len-std-dev \
80 --no-update-check -o uhr2 merged/merged.gtf star_SAM/uhr2.bam
```
